```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")


plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")

plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
```

```python
x_axis=("Jan", "Feb", "Mar", "April", "May", "June", "July","Aug","Sept", "Oct", "Nov", "Dec")
y_axis=( 10.02, 23.24, 39.20,35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71,20.09)

# Create the plot.
plt.bar(x_axis, y_axis, color="skyblue", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.ylim(0,50)
# Add the legend.
plt.legend()


#Add error bars.
#Add caps to the error bars.
#Change the color of the bars to sky blue.
#Rotate the labels on the x-axis to horizontal.
#Set the y-axis increment to every $5.
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="skyblue", label='Boston')


# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.yticks(np.arange(0, 51, step=5.0))
plt.xticks(rotation =1)

# Add the legend.
plt.legend()
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```
